# Model Experiments

```python
import numpy as np
from sklearn.model_selection import TimeSeriesSplit
```

```python
df = pd.read_csv('Cleaned Datasets/merged_c02_movement_attributes.csv')
```

```python
df['date'] = pd.to_datetime(df['date'])

def doy(date):
    new_year_day = pd.Timestamp(year=date.year, month=1, day=1)
    day_of_the_year = (date - new_year_day).days + 1
    return day_of_the_year
```

```python
df['dayofyear'] = [doy(date) for date in df['date']]
```

```python
df = df.sort_values('dayofyear').reset_index()
```

```python
pd.DataFrame(df.columns.tolist())
```

```python
ml_df = df.iloc[:, [29,3,4,5,6,7,8,15,19,21,22,28,12]]
```

```python
ml_df
```

```python
train_set, test_set= np.split(ml_df, [int(.67 *len(ml_df))])
```

```python
X_train = train_set.iloc[:,0:12]
y_train = train_set.iloc[:,12]
```

```python
X_test = test_set.iloc[:,0:12]
y_test = test_set.iloc[:,12]
```

```python
from sklearn.ensemble.forest import RandomForestRegressor
from sklearn.metrics import r2_score
RF_Model = RandomForestRegressor(n_estimators=50,
                                 max_features=1, oob_score=True)
```

```python
rgr = RF_Model.fit(X_train, y_train)
```

```python
y_preds = rgr.predict(X_test)
```

```python
r2_score(y_test,y_preds)
```
